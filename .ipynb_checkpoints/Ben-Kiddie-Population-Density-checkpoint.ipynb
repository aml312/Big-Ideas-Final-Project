# Best Pittsburgh Neighborhood by Livability
Our group came decided to investigate which neighborhood in Pittsburgh was the best by using the metric of livability. There are a number of different submetrics That can be used to measure this, but the on I chose to investigate was population density.

## Why population density?
For the purposes of this project, our group defined livability as the conditions which contribute to an increased quality of life for the people within a certain neighborhood. By this definition, population density can be viewed as a contributing factor to livability, as a lower population density means that individuals can take up more space within the environment, contributing to greater comfort in public spaces. In other words, the lower the population density of a region, the more comfortable that area will generally be to be in, the more livable it is. 

## The data
The dataset I used to measure and compare each neighborhood's population density came from the 2010 SNAP Census. Here are the first few rows of data.

```python
import pandas as pd # we're obviously gonna need pandas for this stuff
import geopandas

dataSet = pd.read_excel('population-density.xlsx') # importing data from Excel file
dataSet.head(5) # showing first 5 rows
```

That's all well and good, but there are plenty of unnecessary metrics in this dataset that we don't really care about. Let's filter out everything except the neighborhood, population in 2010, land area, and population density.

```python
popDenData = dataSet[["Neighborhood", "Pop. 2010", "Land Area (sq. mi)", 
                         "Persons / sq. mi (2010)"]]
popDenData.head(5)
```

Now that we're only looking at the relevant information, let's order the data in ascending order by population density.

```python
popDenData.sort_values(by = ["Persons / sq. mi (2010)"], ascending = True)
popDenData.head(5)
```

You'll notice that some of the lowest-scoring neighborhoods have fewer than 100 people living in them. While this doesn't necessarily disqualify them, it does raise some doubts as to the availability of housing. It's likely that these neighborhoods are almost exclusively industrial and not residential. Thus, we will be removing them.

```python
popMask = popDenData['Pop. 2010'] >= 100
finalPopDenData = popDenData[popMask]

finalPopDenData.head(7)
```

We now have our final dataset. Let's plot the top 7 finishers and get a visual idea of what we're working with.

```python
finalPopDenData.head(7).plot.barh( x='Neighborhood', y='Persons / sq. mi (2010)', rot = 0 )
```

We can also make a heat map of population density using our unmasked data. That looks like this.

```python
neighborhoods = geopandas.read_file("Data-Visualization/Neighborhoods/Neighborhoods_.shp") # read in the shapefile

# do the merge
steps_map = neighborhoods.merge(popDenData, how='left', left_on='hood', right_on='neighborhood')
# look at the head to confirm it merged correctly
steps_map[['hood','number_of_steps','geometry']].head()

popDenData.plot(column='Persons / sq. mi (2010)', # set the data to be used for coloring
               cmap='OrRd',              # choose a color palette
               edgecolor="white",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "People per Square Mile"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```
